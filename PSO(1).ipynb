```python
import os
from rdkit import Chem

# Directory with ligand files in sdf format
data_dir = r"C:\PSO\PL merged"

# List all sdf files
files = [f for f in os.listdir(data_dir) if f.endswith('.sdf')]

# Load molecules into RDKit Mol objects
molecules = [Chem.SDMolSupplier(os.path.join(data_dir, file))[0] for file in files]

# Remove None values (failed loading)
molecules = [mol for mol in molecules if mol is not None]

# Print the number of loaded molecules
print("Number of loaded molecules:", len(molecules))
```

```python
#visaulise the first ligand
def show_molecule(mol):
    mol_block = Chem.MolToMolBlock(mol)
    viewer = py3Dmol.view(width=400, height=400)
    viewer.addModel(mol_block, format='mol')
    viewer.setStyle({'stick': {}})
    viewer.zoomTo()
    viewer.show()
show_molecule(molecules[1])
```

```python
#coordiantes of the first ligand as example
mol = molecules[1]  # Take the first molecule

# Iterate over the atoms in the molecule
for atom in mol.GetAtoms():
    idx = atom.GetIdx()
    symbol = atom.GetSymbol()
    position = mol.GetConformer().GetAtomPosition(idx)
    print(f"Atom {idx}: {symbol} at coordinates (x, y, z) = ({position.x:.2f}, {position.y:.2f}, {position.z:.2f})")
```

```python
# Display structural components for the first five ligands
for j, mol in enumerate(molecules[:5]):
    print(f"Summary for molecule {j+1}:")

    atom_count = mol.GetNumAtoms()
    bond_count = mol.GetNumBonds()
    total_charge = sum([atom.GetFormalCharge() for atom in mol.GetAtoms()])

    print(f"  Number of atoms: {atom_count}")
    print(f"  Number of bonds: {bond_count}")
    print(f"  Total formal charge: {total_charge}")

    print("=" * 30) 
```

```python
import os
from rdkit import Chem
from rdkit.Chem import AllChem

# Initialize the list to store converted ligands
converted = []

# Convert and save the first 50 filtered molecules to PDBQT format
for idx, mol in enumerate(filtered_molecules[:50]):
    # Generate 3D coordinates
    mol = Chem.AddHs(mol)
    AllChem.EmbedMolecule(mol, AllChem.ETKDG())
    
    # Convert the molecule to PDBQT format and store in the 'converted' list
    pdbqt_content = Chem.MolToPDBBlock(mol)
    converted.append(pdbqt_content)
    
    print(f"Converted ligand {idx+1} to PDBQT format")

# Define the directory where you want to save PDBQT files
output_dir = r"C:\PSO\PDBQT_files"

# Create the output directory if it doesn't exist
if not os.path.exists(output_dir):
    os.makedirs(output_dir)

# Save the converted molecules to PDBQT files
for idx, pdbqt_content in enumerate(converted):
    # Define the output PDBQT file path
    output_path = os.path.join(output_dir, f"ligand_{idx+1}.pdbqt")
    
    # Save the PDBQT file
    with open(output_path, 'w') as f:
        f.write(pdbqt_content)
    
    print(f"Saved converted ligand {idx+1} to PDBQT file: {output_path}")
```

```python
#docking with receptor
import os

# AutoDock Vina executable path
vina_path = r"C:\Program Files (x86)\The Scripps Research Institute\Vina\vina.exe"

# Ligand PDBQT directory
ligand_dir = r"C:\PSO\PDBQT_files"

# Protein PDB path
protein_pdb_path = r"C:\PSO\8gcy.pdb"

# Convert protein to PDBQT format
protein_pdbqt_path = os.path.join(ligand_dir, "protein.pdbqt")
os.system(f'"{vina_path}" --receptor "{protein_pdb_path}" --ligand "{protein_pdb_path}" --out "{protein_pdbqt_path}" --center_x 0 --center_y 0 --center_z 0 --size_x 20 --size_y 20 --size_z 20')

# Docking for the first 3 ligands
for idx in range(1, 4):
    ligand_path = os.path.join(ligand_dir, f"ligand_{idx}.pdbqt")
    docking_command = f'"{vina_path}" --receptor "{protein_pdbqt_path}" --ligand "{ligand_path}" --center_x 0 --center_y 0 --center_z 0 --size_x 20 --size_y 20 --size_z 20'
    
    # Run the docking command and capture the result
    docking_result = os.popen(docking_command).read()
    
    # Print the docking result for visualization
    print(f"Docking result for ligand {idx}:\n{docking_result}")
```
